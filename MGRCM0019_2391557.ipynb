```python
import pandas as pd
import numpy as np
import datetime as dt

# Set the display format, avoid using Scientific notation, and keep two decimal places
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Load data
trips_q1 = pd.read_csv('Divvy_Trips_2017_Q1.csv')
trips_q2 = pd.read_csv('Divvy_Trips_2017_Q2.csv')
stations = pd.read_csv('Divvy_Stations_2017_Q1Q2.csv')

# Merge Q1 and Q2 data
trips = pd.concat([trips_q1, trips_q2])


# Convert start and end times to date time format
trips['start_time'] = pd.to_datetime(trips['start_time'])
trips['end_time'] = pd.to_datetime(trips['end_time'])

# Calculate the duration of each trip in minutes
trips['duration_minutes'] = trips['tripduration'] / 60

# To calculate the distance, first merge the station data with the itinerary data
stations.set_index('id', inplace=True)
trips = trips.join(stations[['latitude', 'longitude']], on='from_station_id', rsuffix='_start')
trips = trips.join(stations[['latitude', 'longitude']], on='to_station_id', rsuffix='_end')

# Calculate the distance between two points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius (kilometers)
    dLat = np.radians(lat2 - lat1)
    dLon = np.radians(lon2 - lon1)
    a = np.sin(dLat/2) * np.sin(dLat/2) + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dLon/2) * np.sin(dLon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    distance = R * c
    return distance

# Calculate the distance per trip in kilometers
trips['distance_km'] = trips.apply(lambda row: haversine(row['latitude'], row['longitude'], row['latitude_end'], row['longitude_end']), axis=1)

# Analyze the distribution of order duration and cycling distance
duration_distribution = trips['duration_minutes'].describe().rename(index={'count': 'number', 'mean': 'average value', 'std': 'standard deviation', 'min': 'minimum value', '25%': '25% quantile', '50%': 'median', '75%': '75% quantile', 'max': 'Maximum value'})
distance_distribution = trips['distance_km'].describe().rename(index={'count': 'number', 'mean': 'average value', 'std': 'standard deviation', 'min': 'minimum value', '25%': '25% quantile', '50%': 'median', '75%': '75% quantile', 'max': 'Maximum value'})

# Calculate user's age
current_year = dt.datetime.now().year
trips['age'] = current_year - trips['birthyear']

# Analyzing the behavior of different age groups
age_groups = trips.groupby(pd.cut(trips['age'], bins=np.arange(0, 100, 10)))['trip_id'].count().rename('Cycling frequency')
average_duration_per_age_group = trips.groupby(pd.cut(trips['age'], bins=np.arange(0, 100, 10)))['duration_minutes'].mean().rename('Average cycling time')

# Output Results
print("Order duration distribution (minutes)：\n", duration_distribution)
print("Cycling distance distribution (kilometers)：\n", distance_distribution)
print("Cycling frequency of different age groups：\n", age_groups)
print("Average cycling time for different age groups (minutes)：\n", average_duration_per_age_group)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Calculate the number of times each station serves as a starting point and destination
station_usage_start = trips['from_station_id'].value_counts().reset_index()
station_usage_end = trips['to_station_id'].value_counts().reset_index()

station_usage_start.columns = ['station_id', 'Departure Times']
station_usage_end.columns = ['station_id', 'Arrival Times']

station_usage = pd.merge(station_usage_start, station_usage_end, on='station_id')

#Merge site usage frequency data with site location data
station_usage = pd.merge(station_usage, stations, left_on='station_id', right_on='id')

#The frequency and location of visual site usage
plt.figure(figsize=(12, 10))
sns.scatterplot(data=station_usage, x='longitude', y='latitude', size='Departure Times', sizes=(50, 200), alpha=0.5, legend=False)
plt.title('Station Usage and Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

#Output statistical data on site usage frequency to evaluate the rationality of site location
print("Site usage frequency statistics：\n", station_usage[['Departure Times', 'Arrival Times']].describe())

# Analyze areas with high demand and areas with low demand
high_demand = station_usage[station_usage['Departure Times'] + station_usage['Arrival Times'] > station_usage[['Departure Times', 'Arrival Times']].mean().sum()]
low_demand = station_usage[station_usage['Departure Times'] + station_usage['Arrival Times'] < station_usage[['Departure Times', 'Arrival Times']].mean().sum()]

print("Number of high demand sites：", len(high_demand))
print("Number of low demand sites：", len(low_demand))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Convert start time to date format
trips['start_time'] = pd.to_datetime(trips['start_time'])
trips['start_date'] = trips['start_time'].dt.date

# Calculate the average cycling time per day
daily_trends = trips.groupby('start_date').agg({
    'tripduration': 'mean'  # Average cycling time
}).rename(columns={'tripduration': 'Average cycling time (seconds)'})

# Draw a time trend chart
plt.figure(figsize=(14, 7))
plt.plot(daily_trends.index, daily_trends['Average cycling time (seconds)'])
plt.title('Daily Average Trip Duration Trend')
plt.xlabel('Date')
plt.ylabel('Average Trip Duration (s)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Output trend analysis results
print("Trend analysis of daily average cycling time：\n", daily_trends['Average cycling time (seconds)'].describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


stations = pd.read_csv('Divvy_Stations_2017_Q1Q2.csv')

# Merge Q1 and Q2 data
trips = pd.concat([trips_q1, trips_q2])

# Convert start time to date time format and extract date
trips['start_time'] = pd.to_datetime(trips['start_time'])
trips['start_date'] = trips['start_time'].dt.date

# Merge station data into travel data for distance calculation
stations.set_index('id', inplace=True)
trips = trips.join(stations[['latitude', 'longitude']], on='from_station_id', rsuffix='_start')
trips = trips.join(stations[['latitude', 'longitude']], on='to_station_id', rsuffix='_end')

# Calculate the distance between two points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius (kilometers)
    dLat = np.radians(lat2 - lat1)
    dLon = np.radians(lon2 - lon1)
    a = np.sin(dLat/2) * np.sin(dLat/2) + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dLon/2) * np.sin(dLon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    distance = R * c
    return distance

# Calculate the distance for each trip
trips['distance_km'] = trips.apply(lambda row: haversine(row['latitude'], row['longitude'], row['latitude_end'], row['longitude_end']), axis=1)

# Calculate the average daily cycling distance
daily_distance = trips.groupby('start_date')['distance_km'].mean()

# Draw a trend chart of the average daily cycling distance
plt.figure(figsize=(14, 7))
plt.plot(daily_distance.index, daily_distance, label='Average Daily Distance (km)')
plt.title('Daily Average Trip Distance Trend')
plt.xlabel('Date')
plt.ylabel('Average Trip Distance (km)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Output trend analysis results
print("Trend analysis of daily average cycling distance：\n", daily_distance.describe())
```
